# The causes of Gun Violence in America

A study to better understand the use of firearms in America in assaults, homicides, injuries and suicides.
![](https://www.americangrit.com/wp-content/uploads/2016/10/40-Reasons-to-Ban-Guns-1024x640.jpg)
![](http://www.gunworldofsfl.com/images/Images/over90Guns-For-Rent.gif)

```python
Dataset comprised of all US gun violence incidents (injuries and deaths) from Jan 2013 - March 2018.
Contains nearly 240,000 data points along with data on location, number of people involved, type of firearm(s) used.

```

```python
import pandas as pd
gv = pd.read_csv('gun-violence-data.csv', index_col = 0)
gv.info()
```

```python
Non-null data available for dates, city/state but becomes more sparse for gun types, whether gun was stolen, number of guns involved, relationship of participants.
```

```python
gv.describe()
```

```python
#plt.figure()
%matplotlib notebook
import seaborn as sns
import matplotlib.pyplot as plt
states = gv['state'].value_counts()
cities = gv['city_or_county'].value_counts()
plt.figure(figsize=(12,8))
sns.barplot(states.values, states.index)
plt.xlabel("Number of Incidents",fontsize=15)
plt.ylabel("States",fontsize=15)
plt.title("Incidents per State",fontsize=20)
sns.despine(left=True,right=True)
```

```python
#plt.figure()
cities50 = cities[:50]
plt.figure(figsize=(12,8))
sns.barplot(cities50.values, cities50.index)
plt.xlabel("Number of Incidents",fontsize=15)
plt.ylabel("City/County",fontsize=15)
plt.title("Incidents in top 50 City/County",fontsize=20)
sns.despine(left=True,right=True)
```
